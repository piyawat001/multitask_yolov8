```python
from ultralytics import YOLO

# โหลดโมเดลและเริ่ม train
model = YOLO('multitask_model.yaml')
results = model.train()
```

```python
# โหลดโมเดลที่ train แล้ว
trained_model = YOLO('path/to/trained/model.pt')

# ทำนาย
results = trained_model('path/to/image.jpg')

# แสดงผลลัพธ์
results.show()
```

```python
results = model('path/to/new_image.jpg')

# ผลลัพธ์สำหรับ Segmentation
segmentation_results = results[0].masks  # mask สำหรับแต่ละวัตถุที่ตรวจพบ
seg_classes = results[0].names  # ชื่อคลาสที่ตรวจพบ
seg_confidences = results[0].boxes.conf  # ค่าความมั่นใจของแต่ละ mask

# ผลลัพธ์สำหรับ Classification
classification_result = results[0].probs  # ความน่าจะเป็นของแต่ละคลาส
predicted_class = results[0].probs.top1  # คลาสที่มีความน่าจะเป็นสูงสุด
class_confidence = results[0].probs.top1conf  # ค่าความมั่นใจของคลาสที่ทำนาย
```
